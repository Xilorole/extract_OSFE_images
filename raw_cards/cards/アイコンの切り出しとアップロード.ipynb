```python
import selenium
from selenium import webdriver
import time

driver = webdriver.Chrome(executable_path="/Users/natsuki-iwano/Desktop/chromedriver")

driver.get('https://wikiwiki.jp/osfe-wiki-jp/?cmd=attach&pcmd=upload&page=Img')

from pathlib import Path

p = Path("./icon").resolve()

import unicodedata
```

```python
driver.get('https://wikiwiki.jp/osfe-wiki-jp/?cmd=attach&pcmd=upload&page=Img')
for idx,filename in enumerate( list(filter(lambda l : "png" in l, os.listdir("./icon")))):
    print(f"{idx:3d} : {filename}",end = " ... ")
    norm_filename = unicodedata.normalize('NFC', filename)
    
    links = set()
    for element in list(filter(lambda x : "</a>" in  x ,driver.page_source.split("<a "))):
        links |= {element.split("</a>")[0].split(">")[1]}
    
    if norm_filename in links:
        driver.get(f"https://wikiwiki.jp/osfe-wiki-jp/?plugin=attach&pcmd=info&file={norm_filename}&refer=Img")
        driver.find_element_by_id("_p_attach_delete").click()
        driver.find_element_by_class_name("trolling-defence-check").click()
        print("deleted.", end=" uploading -> ")
        time.sleep(3)
    image = str(p/filename)
    input_tag ="//input[starts-with(@id,'_p_attach_file')]"
    driver.find_element_by_xpath(input_tag).send_keys(image)
    driver.find_element_by_name("attach").click()
    time.sleep(2)
    print("done.")
    # break
```
